## 1. Wiederholung: Begriff der Funktion
Eine Funktion ist eine Zuordnung, bei der jedem Element einer Menge \( x \) genau ein Element \( y \) zugeordnet wird. 
Beispiele:
- Lineare Funktion: \( f(x) = 2x + 1 \)
- Quadratische Funktion: \( g(x) = x^2 \)

```python
# Plotten von einfachen Funktionen
import numpy as np
import matplotlib.pyplot as plt

# Funktionen definieren
x = np.linspace(-10, 10, 100)


f = 2 * x + 1
g = x ** 2

# Plot
plt.figure(figsize=(8, 6))
plt.plot(x, f, label='f(x) = 2x + 1')
plt.plot(x, g, label='g(x) = x^2')
plt.axhline(0, color='black', linewidth=0.5, linestyle='--')
plt.axvline(0, color='black', linewidth=0.5, linestyle='--')
plt.title('Graphen von f(x) und g(x)')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid()
plt.show()
```

## 2. Addition und Subtraktion von Funktionen
### Beispiel:
Seien \( f(x) = 2x^2 + x + 1 \) und \( g(x) = -x^2 + 3 \). Die Addition ergibt:
\[ h(x) = f(x) + g(x) = (2x^2 + x + 1) + (-x^2 + 3) = x^2 + x + 4 \].

```python
# Plotten von Addition und Subtraktion
def f(x):
    return 2 * x**2 + x + 1

def g(x):
    return -x**2 + 3

def h_add(x):
    return f(x) + g(x)

def h_sub(x):
    return f(x) - g(x)


x = np.linspace(-10, 10, 100)

plt.figure(figsize=(8, 6))
plt.plot(x, f(x), label='f(x) = 2x^2 + x + 1')
plt.plot(x, g(x), label='g(x) = -x^2 + 3')
plt.plot(x, h_add(x), label='h_add(x) = f(x) + g(x)')
plt.plot(x, h_sub(x), label='h_sub(x) = f(x) - g(x)')
plt.axhline(0, color='black', linewidth=0.5, linestyle='--')
plt.axvline(0, color='black', linewidth=0.5, linestyle='--')
plt.title('Addition und Subtraktion von Funktionen')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid()
plt.show()
```

```python
%matplotlib widget


import numpy as np
import matplotlib.pyplot as plt
import sympy as sp

# Symbolische Definition der Funktion
x = sp.Symbol('x')
f = 2 * x**2 + 1 * x + 1  # Die Funktion nur einmal definieren

# Umwandlung in eine numerische Funktion
f_numeric = sp.lambdify(x, f, 'numpy')  # Erzeugt eine numpy-kompatible Funktion

# Automatische Generierung des Labels
latex_label = sp.latex(f)

# Werte für den Plot
x_vals = np.linspace(-10, 10, 100)
f_vals = f_numeric(x_vals)  # Numerische Auswertung der Funktion

# Plot
# plt.clf()
plt.plot(x_vals, f_vals, label=f"$f(x) = {latex_label}$")
plt.xlabel("$x$")
plt.ylabel("$f(x)$")
plt.title("Automatische Label-Generierung mit einer einzigen Definition")
plt.legend()
plt.grid(True)
plt.show()
```

```python
%matplotlib widget


import numpy as np
import matplotlib.pyplot as plt
import sympy as sp
import ipywidgets as widgets

# Symbolische Definition der Funktion
x = sp.Symbol('x')


# Widgets für die Parameter
a_widget = widgets.FloatText(value=1.0, description='a:', step=0.1)
b_widget = widgets.FloatText(value=0.0, description='b:', step=0.1)
c_widget = widgets.FloatText(value=0.0, description='c:', step=0.1)


# Buttons
plot_button = widgets.Button(description="Graph zeichnen")
clear_button = widgets.Button(description="alle Graphen löschen")


# Funktion für das Plotten
def plot_function(_):
    a = a_widget.value
    b = b_widget.value
    c = c_widget.value

    # Symbolische Funktion und Label
    f = a * x**2 + b * x + c
    f_numeric = sp.lambdify(x, f, 'numpy')
    latex_label = sp.latex(f)

    # Werte für den Plot
    x_vals = np.linspace(-10, 10, 500)
    y_vals = f_numeric(x_vals)

    plt.plot(x_vals, y_vals, label=f"$f(x) = {latex_label}$")
    plt.legend()


# Funktion zum Löschen der Plots
def clear_plots(_):
    plt.clf()
    plt.xlabel("$x$")
    plt.ylabel("$f(x)$")
    plt.title("$f(x)=ax^2+bx+c$")
    plt.grid(True)


# Buttons mit Funktionen verbinden
plot_button.on_click(plot_function)
clear_button.on_click(clear_plots)

# Anzeige der Widgets
controls = widgets.HBox([a_widget, b_widget, c_widget, plot_button, clear_button])

clear_plots(_)
plt.show()
display(controls)
```

```python
%matplotlib widget


import numpy as np
import matplotlib.pyplot as plt
import sympy as sp
import ipywidgets as widgets

# Symbolische Definition der Funktion
x = sp.Symbol('x')


# Widgets für die Parameter
f_widget = widgets.Text(value='x', description='f(x)=')


# Buttons
plot_button = widgets.Button(description="Graph zeichnen")
clear_button = widgets.Button(description="alle Graphen löschen")


# Funktion für das Plotten
def plot_function(_):
    try:
        # Symbolische Funktion und Label
        f = sp.sympify(f_widget.value)
        f_numeric = sp.lambdify(x, f, 'numpy')
        latex_label = sp.latex(f)

        # Werte für den Plot
        x_vals = np.linspace(-10, 10, 500)
        y_vals = f_numeric(x_vals)

        plt.plot(x_vals, y_vals, label=f"$f(x) = {latex_label}$")
        plt.legend()

    except sp.SympifyError:
        # Fehler durch ungültigen Eingabestring
        print(f"Ungültige Funktion: '{f_widget.value}' konnte nicht ausgewertet werden.")
    except Exception as e:
        # Allgemeine Fehlerbehandlung
        print(f"Ein Fehler ist aufgetreten: {e}")



# Funktion zum Löschen der Plots
def clear_plots(_):
    plt.clf()
    plt.xlabel("$x$")
    plt.ylabel("$f(x)$")
    plt.title("$Funktionengraph$")
    plt.grid(True)


# Buttons mit Funktionen verbinden
plot_button.on_click(plot_function)
clear_button.on_click(clear_plots)


# Anzeige der Widgets
controls = widgets.HBox([f_widget, plot_button, clear_button])

clear_plots(_)
plt.show()
display(controls)
```
